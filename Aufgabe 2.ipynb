# Performance Measure P

## Sinnvolle Performance Measures

- Accuracy  
- F1-Score / F-Measure (β = 1)  
- Kappa (Cohen’s Kappa)  
- MCC (Matthews Correlation Coefficient)  

## Nicht sinnvolle Performance Measures

- RMSE (Regressionsmetrik)  
  - Da es eine numerische Zielvariable voraussetzt.  

- MAE (Regressionsmetrik)  
  - Für numerische Zielvariablen geeignet.  

- ROC-AUC  
  - Eigentlich ein gutes Verfahren für Binär Klassifikation allerdings sind 14 Instanzen bei dieser Metrik nicht aussagekräftig.
