# 4. Auswahl der Performance-Measures

In meiner analyse lege ich den Fokus auf Accuracy, Recall, Weighted F1, MCC, Kappa, Confusion Matrix, ROC-AUC da diese dazu beitragen die Klassifikationsleistung bestmöglich zu bewerten. RMSE und MAE sind hierbei ebenfalls auf diesen Daten anwendbar, aber dienen zur analyse von Schwellwertoptimierung oder Abweichungen der Wahrscheinlichkeitsvorhersagen, was im medizinischen und finanziellen Bereich weniger sinnvoll wäre.

# 5. Überprüfen welche Hyperparameter die Modelle haben

# J48

| Parameter | Bedeutung | Wirkung |
|------------|------------|------------|
| -C | Pruning-Stärke | Kleiner Wert -> stärkeres Pruning -> einfacherer Baum |
| -M | Min. Instanzen pro Blatt | Größer -> weniger Overfitting |
| -U | Unpruned Tree | Kein Pruning -> hoher Overfitting-Risiko |
| -R | Reduced Error Pruning | Alternative Pruning-Methode |
| -N | Folds für Reduced Error | Anzahl Folds für REP |
| -B | Binary Splits | Erzwingt binäre Splits |
| -S | Seed | Reproduzierbarkeit |

# SVM

| Parameter | Bedeutung | Wirkung |
|------------|------------|------------|
| -C | Regularisierungsparameter | Höher -> weniger Regularisierung -> Overfitting-Risiko |
| -K | Kernel-Typ | Linear, Poly, RBF |
| -G | RBF-Kernel Parameter | Höher -> komplexere Entscheidungsgrenze |
| -E | Toleranz | Abbruchkriterium |
| -P | Epsilon | Numerische Stabilität |

# Naive Bayes

| Parameter | Bedeutung | Wirkung |
|------------|------------|------------|
| -K | Kernel Estimator | Glättet numerische Verteilungen |
| -D | Discretize | Diskretisiert numerische Attribute |
