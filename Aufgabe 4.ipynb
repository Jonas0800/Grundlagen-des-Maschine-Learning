*RBF networks can be trained in two phases: first, an unsupervised clustering approach is used to train the parameters of the Gaussians—the means and variances—are trained, as in Section 20.3.1. # 7. Support Vector Machines & Ensemble Learning

*Online learning is helpful when the data may be changing rapidly over time. It is also useful for applications that involve a large collection of data that is constantly growing, even if changes are gradual. For example, with a database ofmillions ofWeb images, you wouldn’t want to train, say, a linear regression model on all the data, and then retrain from scratch every time a new image is added. It would be more practical to have an online algorithm that allows images to be added incrementally. For most learning algorithms based on minimizing loss, there is an online version based on minimizing regret. It is a bonus that many of these online algorithms come with guaranteed bounds on regret.* (Russell und Norvig 2010, S. 753)

*A support vector machine (see Section 18.9) with 25,000 support vectors achieved an error rate of 1.1%.* (Russell und Norvig 2010, S. 754)

*A virtual support vector machine starts with a regular SVM and then improves it with a technique that is designed to take advantage of the structure of the problem. Instead of allowing products of all pixel pairs, this approach concentrates on kernels formed from pairs of nearby pixels. It also augments the training set with transformations of the examples, just as LeNet did. A virtual SVM achieved the best error rate recorded to date, 0.56%.* (Russell und Norvig 2010, S. 755)

Welche Agentenkomponenten sind beim Spracherwerb betroffen? 

1. Zustands-/Perzept Aktion (Policy): „Wie reagiere ich?“  
2. Inferenz aus Perzeptsequenzen: aus Lautstrom relevante Strukturen inferieren (Wörter, Phrasen, Sprecherintention)  
3. Aktion Reaktion: „Wenn ich dieses Wort sage, passiert jenes“  
4. (4–6) Utility/Action-Value/Goals: soziale Belohnung (Zuwendung, Erfolg im Kommunizieren), Zielzustände (verstanden werden, Objekt erhalten)

## Perzepte und Aktionen des Säuglings

### Perzepte (Inputs)

1. Akustisch: Lautstrom (Phonetik/Prosodie), Sprecherwechsel, Rhythmus, Betonung
2. Visuell: Mundbewegungen, Gestik, Blickrichtung
3. Sozial: Reaktionen der Bezugsperson
4. Kontext/Interner Zustand: Aufmerksamkeit, Motivation, Bedürfnislage

### Aktionen (Outputs)

1. Vokalisationen: Der Prozess von Schreien zu Wörtern  
2. Soziale Handlungen: Blickkontakt, Zeigen

## Welche Lernarten braucht der Säugling?

1. Supervised learning: Objekte, die von der Bezugsperson benannt werden, werden somit für das Kind trainiert
2. Reinforcement learning: Erfolg/Misserfolg von Kommunikation als Reward/Punishment
3. semi-supervised: Wenige klare Labels die der Säugling erst als unlabelete Inputs  erlernen muss
